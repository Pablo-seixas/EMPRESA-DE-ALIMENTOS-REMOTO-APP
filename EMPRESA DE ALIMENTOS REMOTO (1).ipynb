```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

dados = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Tipo de Comida': ['Hambúrgueres', 'Pizza', 'Comida Japonesa', 'Pratos com Carne', 'Massas'],
    'Prato': ['Hambúrguer Clássico', 'Pizza Margherita', 'Sushi (12 peças)', 'Frango Assado', 'Macarrão à Bolonhesa'],
    'Preço (R$)': [60, 80, 120, 100, 72],
    'Quantidade em Estoque': [100, 80, 50, 60, 70]
})

plt.figure(figsize=(12, 8))

data_inicial = datetime.today()
data_ficticia = pd.date_range(data_inicial, periods=3, freq='M')

produtos = []
lucros = []
vendas = []
devolucoes = []

# Loop sobre cada produto
for _, produto in dados.iterrows():
    # Criar uma sequência de datas fictícias para cada produto
    datas_produto = pd.date_range(data_inicial, periods=3, freq='M')

    # Simular vendas e devoluções aleatórias (altere conforme necessário)
    vendas_produto = np.random.randint(10, 50, size=len(datas_produto))
    devolucoes_produto = np.random.randint(0, 10, size=len(datas_produto))

    # Calcular o lucro como preço vezes quantidade vendida
    lucro_produto = produto['Preço (R$)'] * (vendas_produto - devolucoes_produto)

    # Calcular os totais para cada produto
    total_vendas = vendas_produto.sum()
    total_devolucoes = devolucoes_produto.sum()
    total_lucro = lucro_produto.sum()

    # Adicionar resultados às listas
    produtos.append(produto['Prato'])
    vendas.append(total_vendas)
    devolucoes.append(total_devolucoes)
    lucros.append(total_lucro)

resultados = pd.DataFrame({'Produto': produtos, 'Vendas': vendas, 'Devolucoes': devolucoes, 'Lucros': lucros})

print(resultados)

# Plotar gráfico de barras para os resultados
plt.bar(resultados['Produto'], resultados['Vendas'], label='Vendas')
plt.bar(resultados['Produto'], -resultados['Devolucoes'], label='Devoluções')  # Usar valores negativos para devoluções
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.title('Vendas e Devoluções por Produto')
plt.legend()
plt.show()
```

```python
# Vendas e Devoluções por Produto:

# Este gráfico de barras compara as vendas e devoluções de cada produto.
# Cada barra representa um produto, com a altura indicando a quantidade vendida e a quantidade devolvida.
# Essa visualização é valiosa para avaliar o desempenho de cada item no que diz respeito a vendas e
# para identificar produtos que podem ter taxas mais altas de devolução.
```

```python
# Dados de exemplo
dados = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Tipo de Comida': ['Hambúrgueres', 'Pizza', 'Comida Japonesa', 'Pratos com Carne', 'Massas'],
    'Prato': ['Hambúrguer Clássico', 'Pizza Margherita', 'Sushi (12 peças)', 'Frango Assado', 'Macarrão à Bolonhesa'],
    'Preço (R$)': [60, 80, 120, 100, 72],
    'Quantidade em Estoque': [100, 80, 50, 60, 70]
})

# Adicionar o novo prato "Churrasco"
novo_prato = {'ID': 6, 'Tipo de Comida': 'Pratos com Carne', 'Prato': 'Churrasco', 'Preço (R$)': 150, 'Quantidade em Estoque': 40}
dados = dados.append(novo_prato, ignore_index=True)

# Adicionar a coluna de data
dados['Data'] = pd.date_range(start='2023-01-01', periods=len(dados), freq='2M')

# Criar DataFrame para armazenar a projeção de lucro futuro
projecao_lucro = pd.DataFrame()

# Simular projeção de 6 meses
for i in range(6):
    # Criar uma sequência de datas fictícias para cada produto
    datas_produto = pd.date_range(start=dados['Data'].max() + timedelta(days=1), periods=len(dados), freq='D')

    # Simular vendas aleatórias (altere conforme necessário)
    vendas_produto = np.random.randint(5, 20, size=len(dados))

    # Calcular o lucro como preço vezes quantidade vendida
    for _, produto in dados.iterrows():
        lucro_produto = produto['Preço (R$)'] * vendas_produto
        projecao_lucro = pd.concat([projecao_lucro, pd.DataFrame({'Data': datas_produto, 'Prato': produto['Prato'], 'Lucro': lucro_produto})])

# Agrupar por produto e data para obter o lucro total para cada produto em cada data
projecao_lucro = projecao_lucro.groupby(['Prato', 'Data']).sum().reset_index()

# Exibir resultados
print(projecao_lucro)

# Plotar gráfico de linha para os resultados
plt.figure(figsize=(12, 8))
for produto in dados['Prato'].unique():
    dados_produto = projecao_lucro[projecao_lucro['Prato'] == produto]
    plt.plot(dados_produto['Data'], dados_produto['Lucro'], label=produto)

plt.xlabel('Data')
plt.ylabel('Lucro')
plt.title('Projeção de Lucro Futuro por Produto')
plt.legend()
plt.show()
```

```python
# Projeção de Lucro Futuro por Produto:

# Esse gráfico projeta o lucro futuro para cada produto com base em simulações.
# Cada linha representa um produto, e a trajetória da linha mostra a projeção do lucro ao longo do tempo.
# As simulações podem levar em consideração fatores como popularidade, variação de preço, e estoque.
# Essa visualização ajuda a identificar produtos com maior potencial de lucro e a entender como
# diferentes variáveis impactam o desempenho financeiro ao longo do tempo.
```

```python
# Simular projeção de lucro futuro por produto
for i in range(6):
    # Aumentar a popularidade dos pratos ao longo do tempo (vendas aumentam)
    dados['Popularidade'] = np.random.uniform(1.0, 1.5, size=len(dados)) * dados['Popularidade'].shift(1, fill_value=1)
    
    # Simular vendas com base na popularidade
    vendas_produto = np.random.randint(5, 20, size=len(dados)) * dados['Popularidade']

    # Calcular o lucro como preço vezes quantidade vendida
    for _, produto in dados.iterrows():
        lucro_produto = produto['Preço (R$)'] * vendas_produto
        projecao_lucro = pd.concat([projecao_lucro, pd.DataFrame({'Data': dados['Data'], 'Prato': produto['Prato'], 'Lucro': lucro_produto})])

# Agrupar por produto e data para obter o lucro total para cada produto em cada data
projecao_lucro = projecao_lucro.groupby(['Prato', 'Data']).sum().reset_index()

# Exibir gráfico de projeção de lucro futuro
plt.figure(figsize=(12, 8))
for produto in dados['Prato'].unique():
    dados_produto = projecao_lucro[projecao_lucro['Prato'] == produto]
    plt.plot(dados_produto['Data'], dados_produto['Lucro'], label=produto)

plt.xlabel('Data')
plt.ylabel('Lucro')
plt.title('Projeção de Lucro Futuro por Produto com Aumento de Popularidade')
plt.legend()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# bd
conn = sqlite3.connect('dados_produtos.db')

# Consultar dados do banco de dados
consulta = pd.read_sql_query('SELECT ID, "Tipo de Comida", Prato, "Preço (R$)", "Quantidade em Estoque" FROM produtos', conn)

# Criar um gráfico de rede
G = nx.Graph()

# Adicionar nós e arestas com base no ID
for _, produto in consulta.iterrows():
    G.add_node(produto['ID'], label=produto['Prato'], tipo=produto['Tipo de Comida'], preco=produto['Preço (R$)'], estoque=produto['Quantidade em Estoque'])
    if produto['ID'] > 1:
        G.add_edge(produto['ID'], produto['ID'] - 1)  # Conectar com o nó anterior

# Desenhar o gráfico de rede
pos = nx.spring_layout(G)  # Layout do gráfico
labels = nx.get_edge_attributes(G, 'weight')
node_labels = nx.get_node_attributes(G, 'label')
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='skyblue')
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_weight='bold', font_size=8)

plt.title('Rede de Produtos')


plt.show()

conn.close()
```

```python
# Rede de Produtos:

#Este gráfico representa uma rede de produtos, onde cada nó representa um produto único identificado pelo seu ID. Os nós são conectados, formando uma rede que destaca a ordem de entrada dos produtos. Cada nó possui informações valiosas, como o tipo de comida, o preço e a quantidade em estoque. Observar a conexão entre os produtos pode fornecer insights sobre padrões de lançamento e estratégias de precificação ao longo do tempo. Esta visualização dinâmica é uma ferramenta poderosa para analisar a evolução do portfólio de produtos e tomar decisões informadas para impulsionar o crescimento e a eficiência operacional.
```
